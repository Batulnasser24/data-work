```python
import pandas as pd
from pathlib import Path



def find_root():
    path = Path().resolve()
    for parent in [path] + list(path.parents):
        if (parent / "data").exists():
            return parent
    return path

ROOT = find_root()

file_path = ROOT / "data" / "processed" / "orders_clean.parquet"


if file_path.exists():
    df = pd.read_parquet(file_path)
    print( {len(df)})
    
    display(df.head())
```
