```python
import pandas as pd
from pathlib import Path

def get_project_root():
    curr = Path().resolve()
    for parent in [curr] + list(curr.parents):
        if (parent / "data").exists():
            return parent
    return curr

ROOT_DIR = get_project_root()

FILE_PATH = ROOT_DIR / "data" / "processed" / "orders_clean.parquet"

if FILE_PATH.exists():
    df = pd.read_parquet(FILE_PATH)
    print( {df.shape})
    display(df.head())
```
