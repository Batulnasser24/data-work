```python
import pandas as pd
from pathlib import Path

# 1. تحديد مكان المجلد الرئيسي للمشروع (data-work)
# سيبحث الكود عن المجلد الذي يحتوي على مجلد 'data' تلقائياً
def find_root():
    path = Path().resolve()
    for parent in [path] + list(path.parents):
        if (parent / "data").exists():
            return parent
    return path

ROOT = find_root()

# 2. تحديد مسار ملف orders.parquet
# تأكدي إذا كان الملف في مجلد processed أو raw
file_path = ROOT / "data" / "processed" / "orders_clean.parquet"

# 3. قراءة الملف وعرض البيانات
if file_path.exists():
    df = pd.read_parquet(file_path)
    print(f"✅ Loaded successfully! Total rows: {len(df)}")
    
    # عرض أول 5 صفوف بتنسيق النوت بوك الجميل
    display(df.head())
else:
    print(f"❌ File not found at: {file_path}")
    print("Check if you ran the processing scripts first!")
```
