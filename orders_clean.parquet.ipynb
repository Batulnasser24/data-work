```python
import pandas as pd
from pathlib import Path

# 1. البحث عن مجلد المشروع الرئيسي 'data-work'
def get_project_root():
    curr = Path().resolve()
    # يصعد للأعلى حتى يجد المجلد الذي يحتوي على مجلد 'data'
    for parent in [curr] + list(curr.parents):
        if (parent / "data").exists():
            return parent
    return curr

ROOT_DIR = get_project_root()

# 2. تحديد مسار الملف بناءً على الجذر
FILE_PATH = ROOT_DIR / "data" / "processed" / "orders_clean.parquet"

# 3. القراءة والعرض
if FILE_PATH.exists():
    df = pd.read_parquet(FILE_PATH)
    print(f"✅ Data loaded! Shape: {df.shape}")
    display(df.head())
else:
    print(f"❌ Could not find the file at: {FILE_PATH}")
```
