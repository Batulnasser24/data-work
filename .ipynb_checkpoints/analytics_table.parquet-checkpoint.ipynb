```python
from pathlib import Path
import pandas as pd

# 1. Get the current directory of the notebook
current_dir = Path().resolve()

# 2. Find the project root (data-work)
# This loop goes up until it finds a folder named 'data-work'
root = current_dir
while root.name != "data-work" and root.parent != root:
    root = root.parent

# 3. Define the data path starting from the root
# Note: Since 'root' is already 'data-work', we don't repeat the name in the path
file_path = root / "data" / "processed" / "analytics_table.parquet"

# 4. Check existence and load
if file_path.exists():
    df = pd.read_parquet(file_path)
    print(f"✅ Success! Loaded table with {df.shape[0]} rows.")
    display(df.head()) # Using display() is better in notebooks
else:
    print(f"❌ File NOT found at: {file_path}")
    print(f"Current working directory: {current_dir}")
```
